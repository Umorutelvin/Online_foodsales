# Selecting features to train

***features need to be transformed before selection to train***

```python
encoder = LabelEncoder()
categorical_features = ['Gender', 'Marital Status', 'Occupation', 'Monthly Income', 'Educational Qualifications', 'Output', 'Unnamed: 12']
for feature in categorical_features:
    of_df[feature] = encoder.fit_transform(of_df[feature])
```

***Training of the features i had to pick 80/20 to to train because the datatset is not that much to create a model,***

```python
x = of_df.drop(['Feedback', 'latitude', 'longitude', 'Pin code'], axis=1)
y = encoder.fit_transform(of_df['Feedback'])


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=5, random_state=42)
rf_model.fit(x_train, y_train)

y_pred = rf_model.predict(x_test)
accuracy_rf = accuracy_score(y_test, y_pred)
report_rf = classification_report(y_test, y_pred)

print(accuracy_rf)
print(report_rf)
print(rf_model)
```

```python
from sklearn.linear_model import LogisticRegression

reg = 0.01

model = LogisticRegression(C=1/reg, solver="liblinear").fit(x_train, y_train)

predictions = model.predict(x_test)
print('Accuracy: ', accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
print (model)
```
